```python
dan=pd.read_excel('Hash-Analytic-Python-Analytics-Problem-case-study-1.xlsx','Existing employees')
dan
```

```python
dan_col = dan['satisfaction_level']
high_satifaction_table = dan[dan_col < 0.5]
high_satisfaction_count = high_satifaction_table['satisfaction_level'].count()
row_count = dan_col.count()
(high_satisfaction_count / row_count) * 100
```

```python
grouped_by_dept = dan['dept'].value_counts()
grouped_by_dept.sum()
```

```python
dan_col = dan['satisfaction_level']
low_satis_table = dan[dan_col < 0.5]
grouped_low_satis_by_dept = low_satis_table['dept'].value_counts()
grouped_low_satis_by_dept.sum()
```

```python
for dept, low_satis_val, total_val in zip(grouped_low_satis_by_dept.index, grouped_low_satis_by_dept, grouped_by_dept):
    percent_low_satis = round((low_satis_val / total_val) * 100, 2)
    print(f'{dept} = {percent_low_satis}%')
```

```python
# choose employees who are prone to leave
# select existing employees with low salaries, high evaluation, no promotion the last 5 years, low and medium salaries
criteria_0 = np.logical_or(dan['salary'] == 'low', dan['salary'] == 'medium')
criteria_1 = np.logical_and(dan['satisfaction_level'] < 0.5, dan['last_evaluation'] >= 0.5)
criteria_2 = dan['promotion_last_5years'] == 0
combined_criteria = np.logical_and(criteria_0, criteria_1)
combined_criteria = np.logical_and(combined_criteria, criteria_2)
prone_to_leave_next = dan[combined_criteria].reset_index()
# save employees who are prone to leave
prone_to_leave_next.to_csv('prone_to_leave_next.csv')
# get number of employees who are prone to leave by department
prone_to_leave_next_by_dept = prone_to_leave_next['dept'].value_counts()
prone_to_leave_next_by_dept
```

```python
# get the toal number of employees who are prone to leave
total_number_prone_to_leave = prone_to_leave_next_by_dept.sum()
```

```python
# get the percentage of employees who are leaving per department
(prone_to_leave_next_by_dept / total_number_prone_to_leave) * 100
```
